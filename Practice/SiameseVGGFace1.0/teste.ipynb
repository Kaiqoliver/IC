```python
from keras.models import  Model
from keras.layers import Layer, Flatten, Dense, Input, Lambda
from keras_vggface.vggface import VGGFace
import keras.backend as K
from sklearn.datasets import fetch_lfw_pairs
```

```python
test_data = fetch_lfw_pairs(subset= 'test', funneled=False, resize=1.0, color=True, slice_=(slice(0, 250), slice(0, 250)))
```

```python
images_test = test_data.pairs
labels_test = test_data.target
```

```python
images_test.shape
```

```python
type(images_test)
```

```python
# Assuming 'test_images' is your numpy array with dimensions (1000, 2, 250, 250, 3)
images_shape = images_test.shape

# Reshape the array to (1000*2, 250, 250, 3) to treat each image as a separate entity
images_reshaped = np.reshape(images_test, (-1, images_shape[2], images_shape[3], images_shape[4]))

# Separate the images into two arrays
first_images = images_reshaped[::2]  # Select every other image starting from the first
second_images = images_reshaped[1::2]  # Select every other image starting from the second

# Now, 'first_images' and 'second_images' are arrays with dimensions (1000, 250, 250, 3)
```

```python
class L1Dist(Layer):
  def __init__(self, **kwargs):
    super().__init__()

  def call(self, input_embedding, validation_embedding):
    return tf.math.abs(input_embedding - validation_embedding)
```

```python
# Reload model
siamese_model = tf.keras.models.load_model('/content/drive/MyDrive/IC/siamesemodelv3.h5',
                                   custom_objects={'L1Dist':L1Dist, 'BinaryCrossentropy':tf.losses.BinaryCrossentropy})
```

```python
batch_size = 16
y_hat = siamese_model.predict([first_images, second_images], batch_size=batch_size)
```

```python
len(y_hat)
```

```python
for i in range(1000):
  if (y_hat[i] > 0.5):
    y_hat[i] = 1.0
  else:
    y_hat[i] = 0
print(y_hat)
```

```python
from tensorflow.keras.metrics import Precision, Recall, Accuracy
```

```python
r = Recall()
r.update_state(labels_test, y_hat)
r.result().numpy()
```

```python
p = Precision()
p.update_state(labels_test, y_hat)
p.result().numpy()
```

```python
a = Accuracy()
a.update_state(labels_test, y_hat)
a.result().numpy()
```
